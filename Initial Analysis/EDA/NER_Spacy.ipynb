Cardinals in a chunk refer to a number and are tagged as the CD. These cardinals occur before cardinals noun. Swapping noun cardinals is useful to put the cardinal before the noun


```python
for ent in doc.ents:
    if ent.label_ == 'CARDINAL':
        print("Case Number:", ent.text)
    elif ent.label_ == 'DATE':
        print("First Date:", ent.text)
```

```python
import spacy

# Load the spaCy model for English
nlp = spacy.load("en_core_web_sm")

# The text containing the case number and title
text = "Case Appeal Number: 13-22-00289-CV\n" \
       "Appellate Court: Thirteenth District Court of Appeals, Texas, Corpus Christi – Edinburg\n" \
       "Appellants: Blaine Crews and Hannah Crews\n" \
       "Appellees: Richard Galvan and Sonia Galvan\n" \
       "The appeal originated from the 445th District Court of Cameron County, Texas.\n" \
       "Opinion of the Court:\n" \
       "The opinion was delivered by Justice Benavides with Justices Longoria and Tijerina sitting on the bench."

# Process the text with spaCy
doc = nlp(text)

# Initialize variables to store case number and title
case_number = None
case_title = None

# Iterate through entities identified by spaCy
for ent in doc.ents:
    if ent.label_ == 'CARDINAL':
        case_number = ent.text
    elif ent.label_ == 'ORG':
        case_title = ent.text

# Print the extracted information
print("Case Number:", case_number)
print("Case Title:", case_title)
```

"displacy" is a module in spaCy that provides a simple and visual way to visualize the syntactic and named entity recognition (NER) annotations produced by spaCy.

```python
from spacy import displacy
displacy.render(doc, style="ent", jupyter=True)
```

We can clearly see that the major named entities have been identified by spacy.

```python
for ent in doc.ents:
    if ent.label_ == 'CARDINAL':
        print("Case Number:", ent.text)
    elif ent.label_ == 'DATE':
        print("Date:", ent.text)
    elif ent.label_ == "PERSON":
        print("Names:", ent.text)
```
